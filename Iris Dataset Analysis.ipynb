The Iris flower has a large number of varieties. To this end, Ronald Fisher first created the Fisher's Iris dataset in 1936 in an attempt to be able to differentiate between three classes of Iris.
We are taking data from these three classes of Iris, Iris Setosa, Iris Versicolour and Iris Virginica, these are very similar in shape compared to other Iris' species such as the Blue Bearded or the Dutch. The data set consists of 50 samples from each class of flower comprising of sepal length in cm, sepal width in cm, petal length in cm and petal width in cm for each class of Iris. 



```python
import pandas as pd
import matplotlib.pyplot as plt
```

In the next line I am assigning the name iris.data to the variable filename

```python
Iris_Dataset = 'iris.data'
```

in the next line, I am assigning the contents of the iris.data file to the dataframe df. I am reading it in as a csv type document using the drop down list in VS code when I type in pd.read it gives me that option.

This is just to ensure I have the copy of a reference for when I am tidying up the project. https://www.geeksforgeeks.org/how-to-read-csv-file-with-pandas-without-header/

```python
df = pd.read_csv(Iris_Dataset, header=None)
df
```

As the dataset had no column headers, I have to insert them into the dataset. I am using 0 to 4 for the numerical columns to make it easier for me later but using Iris as a column title for the flowers. I am saving the list of correct column names in a list called headers. 

This is just to ensure I have the copy of a reference for when I am tidying up the project. https://www.geeksforgeeks.org/how-to-add-header-row-to-a-pandas-dataframe/ also Python Data Analysis with Iris Dataset | Data Science, plotting & graphing, https://www.youtube.com/watch?v=02BFXhPQWHQ

```python
df.set_axis([0, 1, 2, 3, "Iris"], axis=1,inplace=True)
headers = ["sepal length", "sepal width", "petal length", "petal width"]
df
```

To improve the comparison of the data, the four columns of data are shown on this next boxplot

```python
fig, ax = plt.subplots(2, 2, figsize=(8, 6))
A = [df[0][df.Iris == 'Iris-setosa'], df[0][df.Iris == 'Iris-virginica'], df[0][df.Iris == 'Iris-versicolor']]
B = [df[1][df.Iris == 'Iris-setosa'], df[1][df.Iris == 'Iris-virginica'], df[1][df.Iris == 'Iris-versicolor']]
C = [df[2][df.Iris == 'Iris-setosa'], df[2][df.Iris == 'Iris-virginica'], df[2][df.Iris == 'Iris-versicolor']]
D = [df[3][df.Iris == 'Iris-setosa'], df[3][df.Iris == 'Iris-virginica'], df[3][df.Iris == 'Iris-versicolor']]

ax[0, 0].boxplot(A, widths = 0.7)
ax[0, 0].set_title(headers[0])
ax[0, 1].boxplot(B, widths = 0.7)
ax[0, 1].set_title(headers[1])
ax[1, 0].boxplot(C, widths = 0.7)
ax[1, 0].set_title(headers[2])
ax[1, 1].boxplot(D, widths = 0.7)
ax[1, 1].set_title(headers[3])

```

```python
fig, ax = plt.subplots()
def set_color(bp):
    plt.setp(bp['boxes'][0], color='blue')
    plt.setp(bp['boxes'][1], color='red')
    plt.setp(bp['boxes'][2], color='green')

A = [df[0][df.Iris == 'Iris-setosa'], df[0][df.Iris == 'Iris-virginica'], df[0][df.Iris == 'Iris-versicolor']]
B = [df[1][df.Iris == 'Iris-setosa'], df[1][df.Iris == 'Iris-virginica'], df[1][df.Iris == 'Iris-versicolor']]
C = [df[2][df.Iris == 'Iris-setosa'], df[2][df.Iris == 'Iris-virginica'], df[2][df.Iris == 'Iris-versicolor']]
D = [df[3][df.Iris == 'Iris-setosa'], df[3][df.Iris == 'Iris-virginica'], df[3][df.Iris == 'Iris-versicolor']]

bp = plt.boxplot(A, positions = [1, 2, 3], widths = 0.7)
set_color(bp)
bp = plt.boxplot(B, positions = [5, 6, 7], widths = 0.7)
set_color(bp)
bp = plt.boxplot(C, positions = [9, 10, 11], widths = 0.7)
set_color(bp)
bp = plt.boxplot(D, positions = [13, 14, 15], widths = 0.7)
set_color(bp)

ax.set_xticks([2, 6, 10, 14])
ax.set_xticklabels(headers)
ax.legend([bp["boxes"][0], bp["boxes"][1],bp["boxes"][2]], ['Iris-setosa', 'Iris-virginica', 'Iris-versicolor'], loc='upper right')
plt.show()
```

Trying to put all boxplots on one graph with labels

```python
fig, ax = plt.subplots()
def set_color(bp):
    plt.setp(bp['boxes'][0], color='blue')
    plt.setp(bp['boxes'][1], color='red')
    plt.setp(bp['boxes'][2], color='green')

A = [df[0][df.Iris == 'Iris-setosa'], df[0][df.Iris == 'Iris-virginica'], df[0][df.Iris == 'Iris-versicolor']]
B = [df[1][df.Iris == 'Iris-setosa'], df[1][df.Iris == 'Iris-virginica'], df[1][df.Iris == 'Iris-versicolor']]
C = [df[2][df.Iris == 'Iris-setosa'], df[2][df.Iris == 'Iris-virginica'], df[2][df.Iris == 'Iris-versicolor']]
D = [df[3][df.Iris == 'Iris-setosa'], df[3][df.Iris == 'Iris-virginica'], df[3][df.Iris == 'Iris-versicolor']]

bp = plt.boxplot(A, positions = [1, 2, 3], widths = 0.7)
set_color(bp)
bp = plt.boxplot(B, positions = [5, 6, 7], widths = 0.7)
set_color(bp)
bp = plt.boxplot(C, positions = [9, 10, 11], widths = 0.7)
set_color(bp)
bp = plt.boxplot(D, positions = [13, 14, 15], widths = 0.7)
set_color(bp)

ax.set_xticks([2, 6, 10, 14])
ax.set_xticklabels(headers)
ax.legend([bp["boxes"][0], bp["boxes"][1],bp["boxes"][2]], ['Iris-setosa', 'Iris-virginica', 'Iris-versicolor'], loc='upper right')
plt.show()
```
