```python
import pandas as pd
import matplotlib.pyplot as plt
```

In the next line I am assigning the name iris.data to the variable filename

```python
filename = 'iris.data'
```

in the next line, I am assigning the contents of the iris.data file to the dataframe df. I am reading it in as a csv type document from the auto prompt in VS code.

https://www.geeksforgeeks.org/how-to-read-csv-file-with-pandas-without-header/

```python
df = pd.read_csv(filename, header=None)
df
```

As the dataset had no column headers, I have to insert them into the dataset. I am using 0 to 4 for the numerical columns to amake it easier for me but I am saving the list of correct column names in a list called headers. 

https://www.geeksforgeeks.org/how-to-add-header-row-to-a-pandas-dataframe/ also Python Data Analysis with Iris Dataset | Data Science, plotting & graphing, https://www.youtube.com/watch?v=02BFXhPQWHQ

```python
df.set_axis([0, 1, 2, 3, "Iris"], axis=1,inplace=True)
headers = ["sepal length", "sepal width", "petal length", "petal width"]
df
```

using the command describe, I can see the statistical details of the data, but this is across all three types of iris

```python
print(df.describe())
```

I am checking how many rows for each species 

FROM Python Data Analysis with Iris Dataset | Data Science, plotting & graphing, https://www.youtube.com/watch?v=02BFXhPQWHQ

```python
df.Iris.value_counts()
```

show a histogram of all the values

```python
plt.hist([df[0], df[1], df[2], df[3]])
```

show a histogram of just the sepal length values

```python
plt.hist(df[0])
```

printing a histogram of all the values but his time adding the labels to the chart. I am using the headers list I created earlier to add the labels

```python
plt.hist([df[0], df[1], df[2], df[3]], label=[headers[0], headers[1], headers[2], headers[3]])
plt.legend()
```

printing out the first two columns of data to compare and adjusting the transparency so that no data is hidden

```python
plt.hist(df[0], alpha = 1)
plt.hist(df[1], alpha = 0.5)
```

printing out the second two columns of data and the transparency shows the marked difference in data

```python
plt.hist(df[2], alpha = 1)
plt.hist(df[3], alpha = 0.5)
```

printing out the 4 columms of data as separate histograms to give a better view of the data

```python
fig, ax=plt.subplots(2, 2, figsize=(8, 4))
ax[0,0].hist(df[0])
ax[0,1].hist(df[1])
ax[1,0].hist(df[2])
ax[1,1].hist(df[3])
ax[0,0].set_title(headers[0])
ax[0,1].set_title(headers[1])
ax[1,0].set_title(headers[2])
ax[1,1].set_title(headers[3])
```

a scatter plot of the first two columns of data showing the correlation between the three types of Iris

```python
colour = {'Iris-setosa':'red', 'Iris-virginica':'green', 'Iris-versicolor':'blue'}
plt.scatter(df[0], df[1], c=df['Iris'].map(colour))
```

The same scatter plot of the second two columns of data showing the correlation between the three types of Iris, this appears much more linear

```python
colour = {'Iris-setosa':'red', 'Iris-virginica':'green', 'Iris-versicolor':'blue'}
plt.scatter(df[2], df[3], c=df['Iris'].map(colour))
```
